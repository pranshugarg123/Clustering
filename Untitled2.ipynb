```python
# Comparative Performance Study of Clustering Algorithms
# Using Iris Dataset from UCI

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Create a DataFrame for better visualization
df = pd.DataFrame(X, columns=feature_names)
df['Species'] = y
df['Species'] = df['Species'].map({0: target_names[0], 1: target_names[1], 2: target_names[2]})

# Display dataset info
print("Dataset Info:")
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print("\nFirst 5 rows:")
print(df.head())

# Preprocessing techniques
preprocessors = {
    'Raw': None,
    'StandardScaler': StandardScaler(),
    'MinMaxScaler': MinMaxScaler(),
    'RobustScaler': RobustScaler()
}

# Clustering algorithms
algorithms = {
    'KMeans': KMeans,
    'Agglomerative': AgglomerativeClustering,
    'DBSCAN': DBSCAN
}

# Function to apply clustering and evaluate
def evaluate_clustering(X_preprocessed, algorithm, algorithm_name, preprocessor_name, k=None):
    if algorithm_name == 'DBSCAN':
        # DBSCAN doesn't require number of clusters as input
        model = algorithm(eps=0.5, min_samples=5)
        labels = model.fit_predict(X_preprocessed)
        n_clusters = len(np.unique(labels[labels != -1]))  # -1 is noise in DBSCAN
    else:
        model = algorithm(n_clusters=k)
        labels = model.fit_predict(X_preprocessed)
        n_clusters = k

    if len(np.unique(labels)) < 2:
        # Need at least 2 clusters for metrics
        return None

    metrics = {
        'Algorithm': algorithm_name,
        'Preprocessing': preprocessor_name,
        'K': n_clusters,
        'Silhouette': silhouette_score(X_preprocessed, labels),
        'Davies-Bouldin': davies_bouldin_score(X_preprocessed, labels),
        'Calinski-Harabasz': calinski_harabasz_score(X_preprocessed, labels)
    }

    return metrics

# Evaluate all combinations
results = []

# Test different numbers of clusters for KMeans and Agglomerative
for k in [2, 3, 4]:
    for preprocessor_name, preprocessor in preprocessors.items():
        if preprocessor:
            X_preprocessed = preprocessor.fit_transform(X)
        else:
            X_preprocessed = X.copy()

        # KMeans
        km_metrics = evaluate_clustering(X_preprocessed, KMeans, 'KMeans', preprocessor_name, k)
        if km_metrics:
            results.append(km_metrics)

        # Agglomerative
        agg_metrics = evaluate_clustering(X_preprocessed, AgglomerativeClustering, 'Agglomerative', preprocessor_name, k)
        if agg_metrics:
            results.append(agg_metrics)

# DBSCAN (doesn't use k parameter)
for preprocessor_name, preprocessor in preprocessors.items():
    if preprocessor:
        X_preprocessed = preprocessor.fit_transform(X)
    else:
        X_preprocessed = X.copy()

    dbscan_metrics = evaluate_clustering(X_preprocessed, DBSCAN, 'DBSCAN', preprocessor_name)
    if dbscan_metrics:
        results.append(dbscan_metrics)

# Create results DataFrame
results_df = pd.DataFrame(results)

# Display all results
print("\nAll Results:")
print(results_df)

# Filter for best k=3 results (since Iris has 3 true classes)
best_results = results_df[results_df['K'] == 3].sort_values('Silhouette', ascending=False)
print("\nBest Results (k=3):")
print(best_results)

# Visualization
# Reduce to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(StandardScaler().fit_transform(X))

# Plot true labels
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title("True Labels")

# Plot best clustering result
best_model = KMeans(n_clusters=3)
best_labels = best_model.fit_predict(StandardScaler().fit_transform(X))
plt.subplot(1, 3, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=best_labels, cmap='viridis')
plt.title("Best Clustering (KMeans with StandardScaler)")

# Plot worst clustering result
worst_model = DBSCAN(eps=0.5, min_samples=5)
worst_labels = worst_model.fit_predict(X)  # No preprocessing
plt.subplot(1, 3, 3)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=worst_labels, cmap='viridis')
plt.title("Worst Clustering (DBSCAN with Raw data)")

plt.tight_layout()
plt.show()

# Metrics comparison plots
plt.figure(figsize=(15, 5))

# Silhouette Score
plt.subplot(1, 3, 1)
sns.barplot(x='Algorithm', y='Silhouette', hue='Preprocessing', data=best_results)
plt.title("Silhouette Score Comparison (k=3)")
plt.ylim(0, 0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Davies-Bouldin Index (lower is better)
plt.subplot(1, 3, 2)
sns.barplot(x='Algorithm', y='Davies-Bouldin', hue='Preprocessing', data=best_results)
plt.title("Davies-Bouldin Index Comparison (k=3)")
plt.ylim(0.5, 1.0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Calinski-Harabasz Index
plt.subplot(1, 3, 3)
sns.barplot(x='Algorithm', y='Calinski-Harabasz', hue='Preprocessing', data=best_results)
plt.title("Calinski-Harabasz Index Comparison (k=3)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# Save results to CSV
results_df.to_csv('clustering_results.csv', index=False)
print("\nResults saved to clustering_results.csv")
```
